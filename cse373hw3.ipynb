```python
class PriorityQ:
    def __init__(self, max_size=1000):
        self.size = 0
        self.heap = [0]*max_size
        self.inv_heap = [-1]*max_size
        self.items = [None for _ in range(max_size)]
        self.contains = lambda p: p > max_size or self.inv_heap[p] != -1
        self.size = 0
        
    def push(self, x, p):
        self.size += 1
        self.inv_heap[p] = self.size
        self.heap[self.size] = p
        self.items[p] = x
        self.sift_up(self.size)
        return p
    
    def delete(self, p):
        idx = self.inv_heap[p]
        item = self.items[p]
        self.size -= 1
        self.swap(idx, self.size)
        self.sift_up(idx)
        self.sift_down(idx)
        self.items[p] = None
        self.inv_heap[p] = -1
    
    def pop(self):
        to_pop = self.heap[1]
        self.size -= 1
        self.swap(1, self.size)
        self.sift_down(1)
        item = self.items[1]
        self.inv_heap[to_pop] = -1
        self.items[1] = None
        self.heap[self.size+1] = -1
        return to_pop, item
        
    def change_item(self, p, x):
        self.item[p] = x
        self.sift_up(self.inv_heap[p])
        self.sift_down(self.inv_heap[p])
    
    def swap(self, i, j):
        self.heap[i], self.heap[j] = self.heap[j],self.heap[i]
        self.inv_heap[self.heap[i]] = i
        self.inv_heap[self.heap[j]] = j
    
    def min_idx(self):
        return self.heap[1]
    
    def top(self):
        print(self.items[1])
        return (self.heap[1], self.items[self.heap[1]])
        
    def sift_up(self, k):
        while k > 1 and self.greater(k//2, k):
            self.swap(k, k//2)
            k //= 2
            
    def sift_down(self, k):
        while 2*k <= self.size:
            j = 2*k
            if j < self.size and self.greater(j, j+1):
                 j += 1
            if not self.greater(k,j):
                break
            self.swap(k, j)
            k = j
    def greater(self, i, j):
        if self.items[self.heap[i]] == None:
            return False
        elif self.items[self.heap[j]] == None:
            return True
        return self.items[self.heap[i]] > self.items[self.heap[j]]
```
